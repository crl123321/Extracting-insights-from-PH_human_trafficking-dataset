```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
traf = pd.read_csv('The Global Dataset 14 Apr 2020.csv')
traf.head()
```

```python
traf.drop(columns = 'By using this data you agree to the Terms of Use: https://www.ctdatacollaborative.org/terms-use', inplace=True)
traf.replace('-99', np.nan, inplace=True)
traf.replace(-99, np.nan, inplace=True)
traf.head()
```

```python
traf['citizenship'].unique()
```

```python
ph = traf[traf['citizenship']=='PH']
ph.head()
```

```python
ph.shape
```

```python
ph['yearOfRegistration'].unique()
```

```python
ph.pivot_table(columns=['yearOfRegistration'], aggfunc='size')
```

```python
ph.pivot_table(columns=['yearOfRegistration'], aggfunc='size').plot(kind='line', x = 'yearOfRegistration', rot=45)
```

```python
ph['CountryOfExploitation'].unique()
```

```python
ph.pivot_table(columns=['CountryOfExploitation'], aggfunc='size')
```

```python
ph.pivot_table(columns=['CountryOfExploitation'], aggfunc='size').plot(kind='bar')
```

```python
means = ph[
          ['meansOfControlDebtBondage',
            'meansOfControlTakesEarnings', 
            'meansOfControlRestrictsFinancialAccess', 
            'meansOfControlThreats', 
            'meansOfControlPsychologicalAbuse', 
            'meansOfControlPhysicalAbuse', 
            'meansOfControlSexualAbuse', 
            'meansOfControlFalsePromises', 
            'meansOfControlPsychoactiveSubstances', 
            'meansOfControlRestrictsMovement', 
            'meansOfControlRestrictsMedicalCare', 
            'meansOfControlExcessiveWorkingHours', 
            'meansOfControlUsesChildren', 
            'meansOfControlThreatOfLawEnforcement', 
            'meansOfControlWithholdsNecessities', 
            'meansOfControlWithholdsDocuments', 
            'meansOfControlOther', 
            'meansOfControlNotSpecified', 
            'meansOfControlConcatenated' ]
          ].sum(axis=0)
means
```

```python
gender = ph.pivot_table(columns=['gender'], aggfunc='size')
age = count = ph.pivot_table(columns=['ageBroad'], aggfunc='size')

fig = plt.figure(figsize = (12, 8))

#plot 1 
plt.subplot(1, 2, 1)
plt.pie(gender, labels=['Female','Male'], autopct='%1.2f%%')
plt.title("Gender Distribution")

#plot 2
plt.subplot(1, 2, 2)
plt.pie(age, labels=['18-20', '30-38', '39-47'], autopct='%1.2f%%')
plt.title("Age Distribution")

plt.show()
```
